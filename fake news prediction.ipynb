data preprocessing

here we are using porterstemmer,this is used for stemming purpose.

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(message)):
  print(i)
  review=re.sub('[^a-zA-Z]'," ", message['title'][i]) # all splcharacters are remved and replacing with blank
  review=review.lower() # all word in lower case
  review=review.split() # words by words,
# i am traversing it usin list compres, if that word not present in stop words then aply stemmin
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
sen_length=20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sen_length)
```

creating model with lastm layer and embedding layer

```python
embedding_vectore_features=40 # every index has 40 features, with different values explained by cbow
model=Sequential()
model.add(Embedding(voc_size,embedding_vectore_features,input_length=sen_length))
model.add(Dropout(0.3))
model.add(LSTM(100)) # lstm with 100,200,300 neurons
model.add(Dropout(0.3))

model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```
